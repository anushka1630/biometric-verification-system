```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_1_embedding = np.random.rand(1, 128)
user_2_embedding = np.random.rand(1, 128)

print("User 1 shape:", user_1_embedding.shape)
print("User 2 shape:", user_2_embedding.shape)
```

```python
similarity_score = cosine_similarity(user_1_embedding, user_2_embedding)[0][0]
threshold = 0.8

print("Similarity Score:", similarity_score)

if similarity_score >= threshold:
    print("User Verified")
else:
    print("Verification Failed")
```
